## Analysis 1 - Comparison of GDP vs Contributors to GDP

```python
# Analysis 1 : Comparison of GDP vs contributors to GDP

i='India' # Coountry for which analysis is to be performed
year=np.array(gdp.loc[i].index).astype(np.int32)
fig,ax=plt.subplots(figsize=(20,10))
plt.xticks(year,rotation=90)
plt.title('India GDP Analysis')

ax.set_ylabel('GDP',color='g',fontsize=12)
ax.set_xlabel('Year',color='b',fontsize=12)
ax2=ax.twinx()
ax2.set_ylabel('GDP of Agri, Ind, Service, Exp, Imp',color='b',fontsize=12)


ax.plot(year,gdp.loc[i],color='g',label='India_GDP',marker='*')
ax2.plot(year,agri.loc[i],color='b',label='Agricutulre_GDP',marker='o')
ax2.plot(year,ind.loc[i],color='r',label='Industry_GDP',marker='o')
ax2.plot(year,service.loc[i],color='grey',label='Service GDP',marker='o')
ax2.plot(year,imp.loc[i],color='black',label='Import_GDP',marker='o')
ax2.plot(year,exp.loc[i],color='orange',label='Export_GDP',marker='o')
ax2.plot(year,final_consumption.loc[i],color='blue',label='Final_Consumption',marker='*')

plt.legend()
```
